These data were obtained from the NASA Langley Research Center (LaRC) POWER Project funded through the NASA Earth Science/Applied Science Program and extracted using NASA's power access API. These data are extracted with respect to meteorological stations in Nepal. Data is extracted for 93 weather stations spanning 62 districts in Nepal. Although I didn't extracted the dataset it was available in Kaggle.com.

The dataset on Nepal's climate data aligns with the United Nations Sustainable Development Goal (UNSDG) 13: Climate Action. It provides detailed meteorological data from 93 weather stations across Nepal. This data can be used to analyze trends in temperature, precipitation, and other climatic parameters, which is vital for monitoring climate change and its effects in Nepal. Since it is a valuable resource for researching and understanding the climate impacts in nepal contributing directly to achieving SDG 13

Parameters:

Date: The date of the recorded climate data.

District: The geographical region within Nepal for which the climate data is recorded. There are 62 districts in this dataset.

Latitude and Longitude: The geographical coordinates of the district location, specifying its position on the Earth's surface.

Precip: Precipitation, measured in millimeters per day (mm/day), represents the amount of rainfall or snowfall.

Pressure: Surface pressure, measured in kilopascals (kPa), represents the atmospheric pressure at the Earth's surface.

Humidity_2m: Specific Humidity at 2 meters above the ground, measured in grams per kilogram (g/kg).

RH_2m: Relative Humidity at 2 meters above the ground, measured as a percentage (%).

Temp_2m: Temperature at 2 meters above the ground, measured in degrees Celsius (°C).

WetBulbTemp_2m: Wet Bulb Temperature at 2 meters above the ground, measured in degrees Celsius (°C).

MaxTemp_2m: Maximum Temperature at 2 meters above the ground, measured in degrees Celsius (°C).

MinTemp_2m: Minimum Temperature at 2 meters above the ground, measured in degrees Celsius (°C).

TempRange_2m: Temperature Range at 2 meters above the ground, measured in degrees Celsius (°C).

EarthSkinTemp: Earth Skin Temperature, measured in degrees Celsius (°C).

WindSpeed_10m: Wind Speed at 10 meters above the ground, measured in meters per second (m/s).

MaxWindSpeed_10m: Maximum Wind Speed at 10 meters above the ground, measured in meters per second (m/s).

MinWindSpeed_10m: Minimum Wind Speed at 10 meters above the ground, measured in meters per second (m/s).

WindSpeedRange_10m: Wind Speed Range at 10 meters above the ground, measured in meters per second (m/s).

WindSpeed_50m: Wind Speed at 50 meters above the ground, measured in meters per second (m/s).

MaxWindSpeed_50m: Maximum Wind Speed at 50 meters above the ground, measured in meters per second (m/s).

MinWindSpeed_50m: Minimum Wind Speed at 50 meters above the ground, measured in meters per second (m/s).

WindSpeedRange_50m: Wind Speed Range at 50 meters above the ground, measured in meters per second (m/s).

This dataset provides a comprehensive set of climate parameters for each day across the 62 districts of Nepal, making it valuable for climate analysis and research.

Correlation Analysis

```python

corre = df.corr()
plt.figure(figsize=(19, 8))
sns.heatmap(corre, annot=True, fmt=".4f", cmap="coolwarm", linewidths=0.5)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.title("Correlation Heatmap")
plt.show()
```

The strongest correlations with isRainy are with precipitation and humidity-related features:

Precipitation shows a strong positive correlation (0.59), which makes intuitive sense - when it's raining, we obviously measure more precipitation. However, the correlation isn't perfect because precipitation also includes other forms of water like snow or mist.
Humidity_2m has a strong positive correlation (0.65), and RH_2m (Relative Humidity) shows a moderate positive correlation (0.54). This aligns with our understanding that rain requires moisture in the air - higher humidity often precedes and accompanies rainfall.

Temperature-related features show moderate positive correlations:

WetBulbTemp_2m has a correlation of 0.55
The correlation with other temperature measurements (Temp_2m, MaxTemp_2m, MinTemp_2m, EarthSkinTemp) ranges from 0.21 to 0.37
This suggests that while temperature has some relationship with rainfall, it's not as strong a predictor as humidity or precipitation measurements

Wind-related features show very weak correlations:

Most wind speed measurements (at both 10m and 50m heights) show correlations close to zero (ranging from -0.017 to 0.071)
This suggests that wind speeds alone might not be very useful for predicting rainfall.

Here we can see that pressure shows a very weak negative correlation i.e -0.0033 which is unusal cause we relate low pressure with rain this weak correlation might indicate that the relationship between pressure and isRainy is much more complex and non-linear than what simple correlation can capture.

Likewise, there are redundant features such as:
Temp_2m
MaxTemp_2m
MinTemp_2m
EarthSkinTemp

these features shows the correlation of 0.94-0.99 with eachother so we can eleminate some of these features without loosing much information.

Same case with Wind Speed:

There are several redundant wind measurements across different heights:

WindSpeed_10m and WindSpeed_50m show a very strong correlation (0.91)
MaxWindSpeed measurements at both heights correlate at 0.87
MinWindSpeed measurements show a correlation of 0.96.

We could likely eliminate either the 10m or 50m measurements without losing much information

**Precipitation VS Rainy Condition**

```python

# Create figure and axis with larger size for better visibility
plt.figure(figsize=(10, 6))

# Create a scatter plot with some jitter to better see overlapping points
# Adding jitter to isRainy (x-axis) since it's binary (0/1)
plt.scatter(df['isRainy'] + np.random.normal(0, 0.04, len(df)),
            df['Precip'],
            alpha=0.5,  # Add transparency to see overlapping points
            color='navy')

# Customize the plot
plt.title('Precipitation vs Rainy Conditions', fontsize=14, pad=15)
plt.xlabel('Is Rainy (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Precipitation', fontsize=12)

# Set x-axis ticks to show binary nature clearly
plt.xticks([0, 1], ['No Rain', 'Rain'])

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

The scatter plot reveals a clear separation between rainy and non-rainy conditions. Days classified as "No Rain" consistently show near-zero precipitation, forming a tight cluster. In contrast, "Rain" days display a wide range of precipitation values from minimal amounts up to 120 units, with a triangular distribution suggesting light rain is more common than heavy rain. There's a notable transition zone near zero precipitation where the classification between rain and no-rain becomes less distinct. The distribution pattern explains the moderate positive correlation (0.59) we observed earlier between precipitation and isRainy, as the relationship is clear but includes significant variation in precipitation amounts during rainy days.

Relative Humidity(RH_2M) vs isRainy

```python
plt.figure(figsize=(8, 6))

# Box plot for RH_2M vs isRainy
sns.boxplot(x='isRainy', y='RH_2m', data=df, palette="Set2")

# Adding labels and title
plt.xlabel('Rainy (isRainy)')
plt.ylabel('Relative Humidity (RH_2M)')
plt.title('Box Plot of RH_2M vs Rainy (isRainy)')

# Display the plot
plt.show()
```

This plot is a box plot that compares the relative humidity (RH_2m) between rainy and non-rainy conditions. The x-axis represents whether it is rainy (1) or not rainy (0), and the y-axis represents the relative humidity (RH_2M).

For the non-rainy condition (0), the box plot shows:

1.The median relative humidity is around 50%.

2.The interquartile range (IQR) is from approximately 40% to 60%.

3.The whiskers extend from about 20% to 100%.


For the rainy condition (1), the box plot shows:

1.The median relative humidity is around 80%.

2.The IQR is from approximately 70% to 90%.

3.The whiskers extend from about 50% to 100%, with several outliers below 50%.



This plot is interesting because it visually demonstrates that relative humidity tends to be higher during rainy conditions compared to non-rainy conditions.
